```python
# lectura de datos en Python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

datos = pd.read_csv('./Datos.csv', encoding = 'ISO-8859-1' )
datos.head()
```

```python
columns_with_nulls = datos.columns[datos.isnull().any()].tolist()

print("Columnas con valores nulos:", columns_with_nulls)
```

```python
# Obtener tipos de datos de cada columna
column_types = datos.dtypes

print("Tipos de datos de cada columna:")
print(column_types)
```

```python
# Rellenar valores nulos en variables de tipo 'object' con una cadena constante
datos['Pais'].fillna('N/A', inplace=True)
datos['Situacion'].fillna('N/A', inplace=True)

columns_with_nulls = datos.columns[datos.isnull().any()].tolist()

print("Columnas con valores nulos:", columns_with_nulls)
```

```python
# Calcular el promedio de cada columna
column_means = datos.mean()

# Reemplazar los valores nulos con los promedios respectivos
datos_filled = datos.fillna(column_means)

# Mostrar el DataFrame después del reemplazo
print("DataFrame después de reemplazar valores nulos:")
print(datos_filled)
```

```python
df = pd.DataFrame(datos)

# Seleccionar automáticamente solo las columnas numéricas
numeric_columns = df.select_dtypes(include=['number']).columns
correlation_matrix = df[numeric_columns].corr()

# Crear un mapa de correlación utilizando Seaborn
plt.figure(figsize=(20, 20))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Mapa de Correlación')
plt.show()
```

```python
#Implementando librerías resumida, recordar que para hacer una regresion multiple se pueden seguir adicionando variables.

import statsmodels.api as sm

X = datos[['Escolarizacion', 'IMC']]
y = datos['Esperanza de vida']

# Añadir la constante al conjunto de características (intercepto)
X = sm.add_constant(X)

# Crear y ajustar el modelo
model = sm.OLS(y, X).fit()

# Imprimir el resumen del modelo
print(model.summary())
```
