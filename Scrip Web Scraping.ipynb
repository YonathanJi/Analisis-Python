```python
from collections import namedtuple

# Crear un namedtuple llamado 'Punto' con los campos 'x' e 'y'
Punto = namedtuple('Punto', ['x', 'y'])

# Crear una instancia de Punto
p = Punto(3, 4)

# Acceder a los campos por nombre
print(f"p.x es: {p.x}")
print(f"p.x es: {p.y}")

# Acceder a los elementos por índice
print(f"p0 es: {p[0]}")
print(f"p0 es: {p[1]}")
```

```python
url = "https://listado.mercadolibre.com.co/computadores#D[A:computadores]"

Articulo = namedtuple('Articulo', ['nombre',
                                   'precio_antes',
                                   'calificacion',
                                  ])
lista_pc = []
```

```python
r = requests.get(url)
html = r.text
html;
```

```python
html_soup = bs(html, 'html.parser')
html_soup;
```

```python
articulos_mercado = html_soup.find_all('li',
                               class_="ui-search-layout__item shops__layout-item ui-search-layout__stack")
articulos_mercado;
```

```python
for articulo in articulos_mercado:
    nombre = articulo.find('h2', class_=["ui-search-item__title"]).text
    
    precio_anterior = None
    precio_anterior_element = articulo.find('div', class_=["ui-search-price__second-line"])
    if precio_anterior_element:
        precio_anterior = articulo.find('span', class_=["andes-money-amount__fraction"]).text
    calificacion = articulo.find('span', class_=["ui-search-reviews__rating-number"])
    
    lista_pc.append(Articulo(nombre, precio_anterior, calificacion))
```

```python
lista_pc;
```

```python
df = pd.DataFrame(lista_pc)
df
```

```python
# URL de la página web a la que vamos a hacer scraping
url = 'https://www.eltiempo.com/'

# Realizar una solicitud HTTP GET a la página web
response = requests.get(url)

html = response.text
# Parsear el contenido HTML de la página web con BeautifulSoup
soup = bs(html, 'html.parser')

soup;

# Encontrar todos los elementos que contienen los títulos de las noticias
etiqueta = "secondary-b1 containt"

noticias = soup.find_all('div',
                         class_=etiqueta)
noticias
```

```python
# En este ejemplo, vamos a extraer la información de los artículos.

soup = bs(str(noticias), 'html.parser')

articles = soup.find_all('article')

for article in articles:
    title = article.find('h2', class_='title-container').text.strip()
    category = article.find('a', class_='category').text
    published_time = article.find('span', class_='published-at').text
    link = article.find('a', class_='title')['href']

    print("Title:", title)
    print("Category:", category)
    print("Published Time:", published_time)
    print("Link:", link)
    print("--------------------------------")
```
