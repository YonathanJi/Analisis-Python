```python
from PIL import Image
import numpy as np
```

```python
# Abre la imagen
img = Image.open("torta.jpg")

# Mostrar la imagen redimensionada
img
```

```python
# convertir la imagen a una matriz NumPy
image_array = np.array(img)
print("El tamaño de la imagen es:" , image_array.shape)
print(image_array)

'''
Altura: 600 píxeles.
Ancho: 600 píxeles.
Canales de color: 3 canales.
Canales (3 - RGB)
'''
```

```python
# obtener las dimensiones de la imagen
ancho, alto = img.size

print(img.size) #alto, ancho
```

```python
gray_scale = np.array   ([
                        [10, 10, 10],
                        [10, 10, 10],
                        [10, 10, 10]
                        ])

tonoRojo = np.array ([
                        [255, 0, 0],
                        [255, 0, 0],
                        [255, 0, 0]
                        ])
print("El tamaño del filtro es:" , tonoRojo.shape)
```

```python
# Multiplicación de matrices
result = np.dot(image_array, gray_scale) #multiplicación de matrices
result_img = Image.fromarray(result.astype(np.uint8)) 

img.show()
result_img.show()
```

```python
# Convierte la imagen a escala de grises
gris = img.convert('L')
gris.show()
```

```python
# Gira la imagen a los grados enviados en rotate
rotate_img = img.rotate(90)
rotate_img.show()
```

```python
# Gira la imagen a blanco y negro
blancoNegro = img.convert("1")
blancoNegro.show()
```

```python
# Guarda la imagen en escala de grises
gris.save("imagen_grises.jpg")

# Guarda la imagen rotada
rotate_img.save("imagen_rotada.jpg")

# Guarda la imagen en blanco y negro
blancoNegro.save("imagen_bn.jpg")
```
